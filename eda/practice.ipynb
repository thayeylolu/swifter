### Merge Column Names That Are the same together

```python

df["Linked To:"] = df["LinkedTo:"].fillna(df[ "Linked To:"])
df['Additional SanctionsInformation'] = df['AdditionalSanctions Information'].fillna(df['Additional SanctionsInformation'])
df['Additional Sanctions Information'] = df['Additional SanctionsInformation'].fillna(df['Additional Sanctions Information'])
df["Aircraft Mode STransponder Code"] = df["Aircraft Mode S Transponder Code"].fillna(df["Aircraft Mode STransponder Code"])
df["Aircraft Serial Identification"] = df["Aircraft SerialIdentification"].fillna(df["Aircraft Serial Identification"])
df["Aircraft Operator"] = df["AircraftOperator"].fillna(df["Aircraft Operator"])
df["Transactions Prohibited For Persons Owned orControlled By U.S. Financial Institutions"] = df["Transactions Prohibited For Persons Owned or Controlled ByU.S. Financial Institutions"].fillna(df["Transactions Prohibited For Persons Owned orControlled By U.S. Financial Institutions"])

# drop "LinkedTo:"
to_drop = ["LinkedTo:","AdditionalSanctions Information" ,
                'Additional SanctionsInformation', 'Aircraft Mode S Transponder Code',
                "Aircraft SerialIdentification", "AircraftOperator",
               "Transactions Prohibited For Persons Owned or Controlled ByU.S. Financial Institutions"
]

df.drop(columns= to_drop , inplace= True)
df.info()
```

```python

df.columns = df.columns.str.replace(" ", "_").str.lower()
df.head()
```

```python
import re
def clean_alt_list(list_):
    list_ = list_.replace("[", '')
    list_ = list_.replace("]", '')
    list_ = list_.replace("['", '[')
    list_ = list_.replace("']", ']')
    list_ = list_.replace(" ", ', ')
    #list_ = list_.replace(r'^', '[')

    return list_

df["program_type"] = df["program_type"].apply(clean_alt_list)

df["program_type"] 
```

```python
fruits_expanded_v1 = df["program_type"].str.split(" ")
#fruits_expanded_v1 = df["program_type"].apply(pd.Series)
fruits_expanded_v1  = dict(fruits_expanded_v1)

program_dataframe = pd.DataFrame.from_dict(fruits_expanded_v1,orient = 'index' )

program_dataframe.columns = ["program_type_" + str(item) for item in  list(program_dataframe.columns)]
```

```python
data = pd.concat([df, program_dataframe], axis=1)
data 
```
